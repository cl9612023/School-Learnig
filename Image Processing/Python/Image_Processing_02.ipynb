```python
#這篇主要是灰階的對比度提升
#載入需要的Library
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
#可以import別的ipynb
import Ipynb_importer
from Image_Processing_01 import graypicture
```

```python
#讀圖片
img = cv2.imread('C:/Users/user/Desktop/7.jpg',cv2.IMREAD_GRAYSCALE)#以灰階方式讀圖片(灰階圖)
w,h = img.shape#圖片的寬w(y方向)和長h(x方向)
#秀圖
plt.imshow(img,cmap="gray")
plt.show()
graypicture(img,w,h)
```

```python
#線性對比度拉伸
def contractstretch(pic,w,h):
    pic = pic.reshape(w*h)#轉換成一維
    pic2 = np.zeros((w*h))#產生新的圖片
    #最大值和最小值
    min1 = 256 
    max1 = -1
    #對應圖片中的每一點進行比較找出最大最小值(可以用python中的max()和min()取代
    for i in range(0,w*h,1):
        if pic[i]>max1:max1=pic[i]
        if pic[i]<min1:min1=pic[i]
    #以相似三角形進行比例拉伸。formulation:[255/(max - min)]*(輸入 - min)
    for i in range(0,w*h,1):
        pic2[i] = int(255*(pic[i]-min1)/(max1-min1))#轉換方程式
    pic2 = pic2.reshape(w,h)#將一維的圖片轉成二維
    return pic2

#執行結果
img2 = contractstretch(img,w,h)
plt.imshow(img2,cmap="gray")
plt.show()
graypicture(img2,w,h)
```
